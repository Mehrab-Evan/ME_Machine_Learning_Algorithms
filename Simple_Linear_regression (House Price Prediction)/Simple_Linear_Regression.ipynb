```python
# This simple liear regression is to predict the house price
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices_simple_linear.csv")
df
```

```python
# Plotting a scatter_plot for getting an idea of distribution
# matplotlib.inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(BDT)')
plt.scatter(df.area, df.price, color="red", marker=".")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price) # Training model using the dataframe
# X axis first then the second y axis which I need to use
```

```python
# Reshape the input to a 2D array
input_value = np.array(2000).reshape(-1, 1)

# Now I can use reg.predict with the reshaped input
predicted_price = reg.predict(input_value)
```

```python
predicted_price
```

```python
input_value = np.array(2400).reshape(-1, 1)
# Now I can use reg.predict with the reshaped input
predicted_price = reg.predict(input_value)
```

```python
# Linear regression uses y = mx + c
# Here m is coeffecient, x is area (independent variable)
# y is dependent variable it's price
# And the c is the intercepts.
# for this problem ---- price = m * area + c
# to get the m or coeffecient I need to do this
reg.coef_
# Linear regression model will calculate this coef automatically
```
